```python
import numpy as np
import pandas as pd
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

2. create a Pivot table and find the total sale amount region wise, manager wise

```python
df.pivot_table(index = ['Region','Manager'], aggfunc = np.sum, values = 'Sale_amt')
```

3. reate a Pivot table and find the total sale amount region wise, manager wise, sales man wise

```python
df.pivot_table(index = ['Region','Manager','SalesMan'], aggfunc = np.sum, values = 'Sale_amt')
```

4.  create a Pivot table and find the item wise unit sold

```python
df.pivot_table(index = 'Item', values = 'Units', aggfunc = 'sum')
```

7. create a Pivot table and count the manager wise sale and mean value of sale amount.

```python
df.pivot_table(index = ['Manager'], aggfunc = [len, 'mean'], values = 'Sale_amt')
```

8. find manager wise, salesman wise total sale and also display the sum of all sale amount at the bottom

```python
df.pivot_table(index = ['Manager', 'SalesMan'], values = ['Units','Sale_amt'], aggfunc = 'sum', fill_value=0,margins=True)
```
