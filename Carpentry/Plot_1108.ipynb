```python
import pandas as pd
data = pd.read_csv('coolhearts.csv')
data.loc[:, 'Area'].plot()
```

```python
data.plot(x = 'Infarcted', y = 'Area')
```

```python
plt.style.use('ggplot')
data.plot(kind = 'hist')
```

```python
import numpy as np
xlab = np.arange(32) # get a seqence 0:31
ylab = data.loc[:, 'Area']
plt.plot(xlab, ylab, 'g--')
```

```python
ylab0 = data.iloc[:,0]
plt.plot(xlab, ylab0, 'r-', label = 'first col')
plt.plot(xlab, ylab, 'g-', label = 'second col')
plt.legend()
plt.savefig('fig1.png')
```

```python
fig2 = plt.gcf()
plt.scatter(xlab, ylab0, s = data.loc[:, 'Group']*5)
fig2.savefig('fig2.png')
```
